```python
import os
import random
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image, ImageEnhance
from skimage import data, img_as_float, exposure
from skimage.util import random_noise

#contrast filter
def change_contrast(img, level):
    factor = (259 * (level + 255)) / (255 * (259 - level))
    def contrast(c):
        return 128 + factor * (c - 128)
    return img

#a=random.randint(0,100)
img = cv2.imread('Hitesh_0.png')
#img_noise = random_noise(img, mode='s&p',amount=0.1)

#img_blur = cv2.bilateralFilter(img,9,75,75)
#img_gaussian_blur = cv2.GaussianBlur(img,(15,15),0)
#img_median_blur = cv2.medianBlur(img,3)
#img_edges = cv2.Canny(img,100,200)
img_contrast = change_contrast(Image.open('Hitesh_0.png'), 100)
img_gamma_corrected = exposure.adjust_gamma(img, 0.5)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_noise = np.array(255*img_noise, dtype = 'uint8')

#bird_blur = cv2.imwrite('bird_blur.png',img_blur)
#bird_gaussian_blur = cv2.imwrite('bird_gaussian_blur.png',img_gaussian_blur)
#bird_median_blur = cv2.imwrite('bird_median_blur.png',img_median_blur)
#bird_edges = cv2.imwrite('bird_edges.png',img_edges)
bird_contrast = img_contrast.save("bird_contrast.png")
bird_gamma_corrected = cv2.imwrite('bird_gamma_corrected.png',img_gamma_corrected)
bird_gray = cv2.imwrite('bird_gray.png',img_gray)
bird_noise = cv2.imwrite('bird_noise.png',img_noise)

#plt.subplot(331),plt.imshow(img_blur),plt.title('img_blur')
#plt.xticks([]), plt.yticks([])
#plt.subplot(332),plt.imshow(img_gaussian_blur),plt.title('img_gaussian_blur')
#plt.xticks([]), plt.yticks([])
#plt.subplot(333),plt.imshow(img_median_blur),plt.title('img_median_blur')
#plt.xticks([]), plt.yticks([])
#plt.subplot(335),plt.imshow(img_edges),plt.title('img_edges')
#plt.xticks([]), plt.yticks([])
plt.subplot(231),plt.imshow(img),plt.title('Original_img')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(img_gamma_corrected),plt.title('img_gamma_corrected')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(img_gray),plt.title('img_gray')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(img_noise),plt.title('img_noise')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(img_contrast),plt.title('img_contrast')
plt.xticks([]), plt.yticks([])

plt.show()
```
